# 04. Frequency distributions and checking for strong outliers

```python
# importing additional libraries
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = df
```

```python
selected_columns = [
    'short_name', 'player_positions', 'overall', 'potential', 'value_eur',
    'wage_eur', 'age', 'height_cm', 'weight_kg', 'preferred_foot',
    'pace', 'shooting', 'passing', 'defending', 'physical'
]
```

```python
# Step 1: Frequency Distribution for Categorical Columns
def frequency_distribution(data, column_name):
    """
    Displays and visualizes the frequency distribution of a categorical column.
    """
    print(f"\nFrequency Distribution for {column_name}:")
    print(data[column_name].value_counts().head(10))  # Display top 10
    plt.figure(figsize=(10, 6))
    sns.countplot(data=data, y=column_name, order=data[column_name].value_counts().index[:10])
    plt.title(f"Top 10 Frequency Distribution of {column_name}")
    plt.xlabel("Count")
    plt.ylabel(column_name)
    plt.show()
```

```python
# Step 2: Detecting Outliers in Numeric Columns
def detect_outliers(data, column_name):
    """
    Detects outliers in a numeric column using the IQR method.
    """
    Q1 = data[column_name].quantile(0.25)
    Q3 = data[column_name].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[column_name] < lower_bound) | (data[column_name] > upper_bound)]
    print(f"\nOutliers in {column_name}: {len(outliers)} rows")
    if not outliers.empty:
        print(outliers[[column_name]].head(5))
    # Boxplot to visualize outliers
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=data[column_name])
    plt.title(f"Boxplot for {column_name} (Outliers Detection)")
    plt.show()
```

```python
# Apply to selected columns
for col in selected_columns:
    if col in data.columns:
        # Frequency Distribution for Categorical Columns
        if data[col].dtype == 'object':
            frequency_distribution(data, col)
        # Outliers Detection for Numeric Columns
        if np.issubdtype(data[col].dtype, np.number):
            detect_outliers(data, col)
```
